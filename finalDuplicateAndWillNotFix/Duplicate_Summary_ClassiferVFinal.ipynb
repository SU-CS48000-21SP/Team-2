```python
from absl import logging

import tensorflow as tf

import tensorflow_hub as hub
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import re
import seaborn as sns

module_url = "https://tfhub.dev/google/universal-sentence-encoder/4"
model = hub.load(module_url)
print ("module %s loaded" % module_url)
def embed(input):
  return model(input)
```

```python
messages = df['Summary'].astype('str').tolist()
message_embeddings = embed(messages)
```

```python
df['Summary'].unique().tolist()
```

```python
from scipy.spatial import distance

def sorter(e):
  return e['value']

def checkDuplicate():
  duplicates = []
  inputSummary = input('summary: ')
  summaryArr = [inputSummary]
  inputEmbedding = embed(summaryArr)
  embeddingSingle = inputEmbedding[0]
  for i, message_embedding in enumerate(np.array(message_embeddings).tolist()):
    cosdistance = (1 - distance.cosine(embeddingSingle, message_embedding))
    if (cosdistance > 0.80):
      duplicates.append({
          'message': 'This summary might be a duplicate of "{}"'.format(messages[i]),
          'value': cosdistance
      })
  duplicates.sort(reverse=True, key=sorter)
  for i in duplicates:
    print(i['message'])
```
